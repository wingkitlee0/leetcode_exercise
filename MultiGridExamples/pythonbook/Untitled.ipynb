```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('figure', dpi=150)

N=16
omega=2.0/3.0
max_its=8

x=np.linspace(0,1,N+1)
s1=np.sin(np.pi*x)
s13=np.sin(13*np.pi*x)

# RHS
f = np.cos(x*np.pi)
# grid spacing
h = x[1]-x[0]

# initial guess
e_old=s1+s13/3.0
```

```python
e=np.zeros_like(e_old)

fig, ax = plt.subplots()
ax.plot(x, e_old, alpha=0.5)
for it in range(max_its):
    e[1:-1]=(1.0-omega)*e_old[1:-1]+\
        0.5*omega*( e_old[0:-2]+e_old[2: ] + h**2 * f[1:-1])
    ax.plot(x, e, alpha=0.5)
    e_old=e.copy()
ax.set_xlabel("The 17 grid points")
ax.set_ylabel("The first 8 iterates of the error")
```
